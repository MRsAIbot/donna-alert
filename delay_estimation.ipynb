```python
LRRID = np.unique(df[['LateRunningReasonID']].dropna(how='any'))
LRRT = np.unique(df[['Reason']].dropna(how='any'))
reason_dict = {int(k):v for k,v in zip(LRRID, LRRT)}
reason_dict
```

```python
df_feats = df.loc[:,['DepartureDelay','LateRunningReasonID','STA','STD','ArrivalTime','DepartureTime']].dropna(how='any')
df_feats.dtypes
```

```python
df_feats[['STA','STD','ArrivalTime','DepartureTime']] = df_feats[['STA','STD','ArrivalTime','DepartureTime']].astype('datetime64[s]')
```

```python
df_feats['ArrivalDelay'] = np.round((df_feats['STA'] - df_feats['ArrivalTime']) / np.timedelta64(1, 's'))
df_feats.dtypes
```
